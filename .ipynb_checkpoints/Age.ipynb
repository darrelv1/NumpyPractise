```python

#Marigold & HVAC Assets 
MAR_HVACdf = pd.read_excel("Marigold Asset HVAC.xlsx", sheet_name="Find Assets")
```

```python
#ALl Marigoolds except: (hvac FF&E amd Suites) Assets
MAR_ALLdf = pd.read_excel("MarigoldAssetAllexpFF&E.xlsx", sheet_name="Find Assets")
```

```python

#All Companies: FF&E Assets 
AmenityFFEdf = pd.read_excel("AmenitySpacesFF&EAsset.xlsx", sheet_name="Find Assets")
```

```python
#Regex utilizing Negative lookahead assertion: x(?!y) for Opening Balance asset naming conventions used in Marigold 
marOBregex = re.compile("[Oo]pening|PPA(?!(\s\w+)?(\s|)?adjust)")

#filters boolean table from the regex: marOBregex
MarOBfilter = MAR_HVACdf["Asset ID"].str.contains( marOBregex ,regex=True)
MarOBALLfilter = MAR_ALLdf["Asset ID"].str.contains( marOBregex ,regex=True)
FFEfilter = AmenityFFEdf["Asset ID"].str.contains(marOBregex, regex=True )
```

```python
#Marigold, HVAC , OB Assets dataframe
MAR_HVACdf = MAR_HVACdf[MarOBfilter]
```

```python
#Marigold All, OB Assets dataframe *except for suites, HVAC and FF&E
MAR_ALLdf = MAR_ALLdf[MarOBALLfilter]
```

```python
#All companies, Ob assets dataframe 
AmenityFFEdf = AmenityFFEdf[FFEfilter]
```

```python
with pd.ExcelWriter("MAR_HVAC.xlsx") as writer:
        MAR_HVACdf.to_excel(writer)
```

```python
from abc import  ABC, abstractmethod
from datetime import datetime
import re


class handlerInterface (ABC):


    @abstractmethod
    def cleanUp(): 
        pass

    @abstractmethod
    def printer():
        pass 
```

```python
class AgeInterface(handlerInterface):

    @classmethod
    def cleanUp(dataframe):
        print(dataframe)
    
    def printer():
        print("this is print")

class example(AgeInterface):

    def __init__(self,dataframe):
        self.dataframe = dataframe
        super().cleanUp(dataframe)
```

```python
example = AgeInterface()
example.cleanUp("Age")
```
